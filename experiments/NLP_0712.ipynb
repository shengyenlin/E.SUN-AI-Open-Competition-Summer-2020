```
stopword = []
with open('/content/gdrive/My Drive/玉山NLP_自己/stopword.txt', 'r', encoding = 'utf8') as file:
  for data in file.readlines():
    data = data.strip()
    stopword.append(data)
```

```
def del_stopword(word_list):
  news_list = word_list.copy()
  for i in range(len(word_list)):
    news_list[i] = []
    for j in range(len(word_list[i])):
      if word_list[i][j] not in stopword:
        news_list[i].append(word_list[i][j])
  return news_list
```

```
news_nocrime_stop = del_stopword(news_nocrime)
news_crime_stop = del_stopword(news_crime)
news_stop = del_stopword(news)
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=CountVectorizer()
transformer=TfidfTransformer()
tfidf=transformer.fit_transform(vectorizer.fit_transform(corpus_crime))
word=vectorizer.get_feature_names()
weight=tfidf.toarray()
```

```
import heapq
#不對的方法
tfidf_list = []
for i in range(len(weight)):
  list_word = [] 
  a = weight[i].tolist()
  max_num_index_list = map((a).index, heapq.nlargest(10, a))
  for j in list(max_num_index_list):
    list_word.append(word[j])
  tfidf_list.append(list_word)
```

```
tfidf_list_10 = []
for i in range(len(weight)):
  list_word = [] 
  a = weight[i].tolist()
  b = sorted(enumerate(a), key=lambda x:x[1], reverse = True)
  index_list = []
  value_list = []
  for index, value in b:
    index_list.append(index)
  for j in index_list[:10]:
    list_word.append(word[j])  
  tfidf_list_10.append(list_word)
```

```
tfidf_list_5 = []
for i in range(len(weight)):
  list_word = [] 
  a = weight[i].tolist()
  b = sorted(enumerate(a), key=lambda x:x[1], reverse = True)
  index_list = []
  value_list = []
  for index, value in b:
    index_list.append(index)
  for j in index_list[:5]:
    list_word.append(word[j])  
  tfidf_list_5.append(list_word)
```
