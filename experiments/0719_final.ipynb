```
entity_all = []
for k in range(len(ner_text2)):
  news_entity = []
  list_ner = sorted(ner_text2[k])
  for i in ws_text2[k]:
    news_entity.append('NONE_PERSON')
    for j in range(len(list_ner)):
      if i == list_ner[j][3] and list_ner[j][2] == 'PERSON':
        news_entity.pop()
        news_entity.append('PERSON')
      else:
        continue
  entity_all.append(news_entity)    
```
